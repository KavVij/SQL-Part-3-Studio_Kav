### **Write a query that returns tag\_id, the number of times each tag\_id is used and the tag\_name.**  

### Use the GROUP BY and HAVING clause to narrow your focus.

Try multiple keywords, ie woman vs female (female ended up giving better results)

```sql
-- Solution
select bt.tag_id,count,tag_name
from booksdb.dbo.book_tags bt 
join  booksdb.dbo.tags as t
on t.tag_id=bt.tag_id
where lower (tag_name) like '%female%' OR 
    lower(tag_name) like '%woman%'

```

### **Write a query to return authors, titles, ratings and** **selected tag(s)\_id.**

```sql
-- Solution
select top 20 title,authors,tag_id,rating
from booksdb.dbo.books as b ,booksdb.dbo.book_tags as t,booksdb.dbo.ratings as r
where t.tag_id=11662 AND 
      b.book_id=t.goodreads_book_id and 
      r.book_id=b.book_id


```

### **Write a query to return authors, titles, ratings and** **selected tag(s)\_id and tag name(s)**

```sql
select top 20 title,authors,t.tag_id,rating,tag_name
from booksdb.dbo.books as b ,booksdb.dbo.book_tags as t,booksdb.dbo.ratings as r,booksdb.dbo.tags as tag
where tag.tag_name='female-writers'  AND 
      b.book_id=t.goodreads_book_id and 
      tag.tag_id=t.tag_id and 
      r.book_id=b.book_id
```

### **A query that returns tag\_id, the number of times each tag\_id is used and the tag\_name.**

```sql
select t.tag_id,count,tag_name
from Booksdb.dbo.book_tags as t
join booksdb.dbo.tags as tag 
on t.tag_id=tag.tag_id
where lower(tag_name) like '%Christmas%'
```

### **A query that returns the kids christmas books,authors and title**

```sql
select t.tag_name,authors,title,book_id
from booksdb.dbo.book_tags as tag
join booksdb.dbo.tags as t 
on t.tag_id=tag.tag_id
join booksdb.dbo.books as b on
   tag.goodreads_book_id= b.book_id
where lower(tag_name) like '%childrens%christmas%'
```
